<p> 
As we know stack follows last in first out, which means we get the most recently inserted element whenever we remove an element from the stack. But queue follows first in first out, it means we get that element which we inserted in the starting at each deletion, it means if we want to use the queue like a stack we have to arrange elements in the queue such that we get the most recent element at each deletion.

Approach:
Take a single queue.
push(x): Push the element in the queue.
Use a for loop of size()-1, remove element from queue and again push back to the queue, hence the most recent element becomes the most former element and vice versa.
pop(): remove the element from the queue.
top(): show the element at the top of the queue.
size(): size of the current queue.
Repeat step3 at every insertion of the element.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
from queue import Queue

class Stack:
    def __init__(self):
        self.q = Queue()


    def push(self, x):
        s = self.q.qsize()
        self.q.put(x)
        for i in range(s):
            self.q.put(self.q.get())


    def pop(self):
        n = self.q.get()
        return n


    def top(self):
        return self.q.queue[0]


    def size(self):
        return self.q.qsize()
```

<p> 
Node Class:
Represents a single element with data (value) and next (pointer to the next node).

MyQueue Class:
front: Points to the first element.
rear: Points to the last element.

push(data):
Create a new node with data.
If the queue is empty (front == None):
Set both front and rear to the new node.
Else:
Link the current rear.next to the new node and update rear.

pop():
If the queue is empty (front == None): Return -1.
Else:
Save front.data, move front to front.next, and return the saved value.
If the queue is now empty, set rear = None.
<br><br>
Time complexity: O(1)<br>
Space Complexity: O(n)</p>

```python
class Node: 
    def __init__(self, data): 
        self.data = data 
        self.next = None
        
class MyQueue:
    def __init__(self):
        self.rear = None
        self.front = None

    def push(self, data): 
        temp = Node(data)
        if self.front == None:
            self.front = temp
            self.rear = temp
        else:
            self.rear.next = temp
            self.rear = temp

    def pop(self):
        if self.front is None:
            return -1
        else:
            res= self.front.data
            self.front = self.front.next
            if self.front== None:
                self.rear==None
            return res
```

<h3>7. Check for balanced paranthesis</h3>
<a href="https://leetcode.com/problems/valid-parentheses/description/">Problem Link</a>
<p> 
Initialize an empty stack.

Traverse each character in the string:
If it’s an opening bracket, push it onto the stack.
If it’s a closing bracket, check if the stack is non-empty and if the top of the stack is the corresponding opening bracket:
If so, pop the opening bracket from the stack.
If not, return false (either there’s a mismatch or no opening bracket to match the closing one).

After traversing the string, check if the stack is empty:
If it is, return true (all brackets are matched).
If it’s not, return false (there are unmatched opening brackets).
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def isValid(self, s: str) -> bool:
        stack = []
        bracket_map = {')': '(', '}': '{', ']': '['}
        for char in s:
            if char in bracket_map:
                top_element = stack.pop() if stack else '#'
                if bracket_map[char] != top_element:
                    return False
            else:
                stack.append(char)
        return not stack
```

<h3>8. Implement Min Stack</h3>
<a href="https://leetcode.com/problems/min-stack/description/">Problem Link</a>
<p> 
In total, need 3 variables - 2 stacks and a min variable.
First stack can be used for the regular operations.
Second stack is used to store the past min values.
The min variable contains the current min value of the first stack.

Push:
Push value to the first stack.
If value being pushed is less than min, push min to second stack and assign value to min.

Pop:
Pop from first stack and store in a temp variable.
If the value popped if same as min, pop a value from second stack and assign it to min given second stack is not empty.
If second stack is empty, assign min as infinity.
<br><br>
Time complexity: O(1)<br>
Space Complexity: O(n)</p>

```python
class MinStack:

    def __init__(self):
        self.stack = []
        self.minStack = []
        self.min = float("inf")
        

    def push(self, val: int) -> None:
        self.stack.append(val)
        if val <= self.min:
            self.minStack.append(self.min)
            self.min = val

    def pop(self) -> None:
        res = self.stack.pop()
        if res == self.min:
            self.min = self.minStack.pop() if len(self.minStack) > 0 else float("inf") 
        

    def top(self) -> int:
        return self.stack[-1]
        

    def getMin(self) -> int:
        return self.min
        


# Your MinStack object will be instantiated and called as such:
# obj = MinStack()
# obj.push(val)
# obj.pop()
# param_3 = obj.top()
# param_4 = obj.getMin()
```

<h3>1. Infix to Postfix Conversion using Stack</h3>
<a href="https://www.geeksforgeeks.org/problems/infix-to-postfix-1587115620/1?utm_source=youtube&utm_medium=collab_striver_ytdescription&utm_campaign=infix-to-postfix">Problem Link</a>
<p> 
The code converts an infix expression (e.g., A+B*(C^D-E)) into its equivalent postfix expression (e.g., ABCD^E-*+) using a stack-based approach. It follows these steps:

Precedence function (prec): Determines the precedence of operators (^, *, /, +, -). Higher value indicates higher precedence.
Main logic: Iterates over each character in the expression:
If the character is an operand (letters or digits), it appends it to the result.
If it is an opening parenthesis (, it pushes it onto the stack.
If it is a closing parenthesis ), it pops elements from the stack to the result until an opening parenthesis is encountered.
If it is an operator, it pops operators from the stack to the result until the top of the stack has an operator with lower precedence.
End of iteration: Appends remaining stack elements to the result.
Finally, it returns the postfix expression.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def InfixtoPostfix(self, expr):
        def prec(c):
            if (c == "^"):
                return 3
            elif (c == "/" or c == "*"):
                return 2
            elif (c == "+" or c == "-"):
                return 1
            else:
                return -1
                
        stack = []
        result = ""
        
        for char in expr:
            if char.isalpha() or char.isdigit():
                result += char
            
            elif char == '(':
                stack.append(char)
            
            elif char == ')':
                while stack and stack[-1] != '(':
                    result += stack.pop()
                stack.pop() 
            
            else:  
                while stack and prec(char) <= prec(stack[-1]):
                    result += stack.pop()
                stack.append(char)
                
        while stack:
            result += stack.pop()

        return result
```

<h3>2. Prefix to Infix Conversion</h3>
<a href="https://www.geeksforgeeks.org/problems/prefix-to-infix-conversion/1?utm_source=youtube&utm_medium=collab_striver_ytdescription&utm_campaign=prefix-to-infix-conversion">Problem Link</a>
<p> 
Easy solution, check comments.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def preToInfix(self, pre_exp):
        stack = []
        
        # Traverse the prefix expression from right to left
        for char in reversed(pre_exp):
            if char.isalpha():  # If the character is an operand
                stack.append(char)
            else:  # If the character is an operator
                operand1 = stack.pop()
                operand2 = stack.pop()
                # Combine the operator and operands in infix format
                stack.append(f"({operand1}{char}{operand2})")
        
        # The final element in the stack is the complete infix expression
        return stack[-1]
```

<h3>1. Next Greater Element I</h3>
<a href="https://leetcode.com/problems/next-greater-element-i/description/">Problem Link</a>
<p> 
Check comments.
<br><br>
Time complexity: O(n+m)<br>
Space Complexity: O(n)</p>

```python
from typing import List
class Solution:
    def nextGreaterElement(self, nums1: List[int], nums2: List[int]) -> List[int]:
        # Stack to keep track of the next greater elements
        stack = []
        # Dictionary to store the next greater element for each number in nums2
        ans = {}
        
        # Traverse nums2 in reverse order
        for i in range(len(nums2) - 1, -1, -1):
            # Remove elements from the stack that are less than or equal to nums2[i]
            while stack and stack[-1] <= nums2[i]:
                stack.pop()
            
            # Determine the next greater element
            if not stack:
                ans[nums2[i]] = -1  # No greater element
            else:
                ans[nums2[i]] = stack[-1]  # Top of the stack is the next greater element
            
            # Push the current element onto the stack
            stack.append(nums2[i])
        
        # Construct the result for nums1 using the ans dictionary
        return [ans[num] for num in nums1]
```

<h3>2. Next Greater Element II
</h3>
<a href="https://leetcode.com/problems/next-greater-element-ii/">Problem Link</a>
<p> 
Doubling the list: nums2 = nums + nums creates a new list, nums2, by concatenating the original list nums with itself. This helps to handle the circular nature of the problem (next greater element wraps around the list).

Stack for tracking elements: The stack s stores elements in decreasing order. It helps to quickly find the next greater element for each number in the list as we iterate through nums2 in reverse.

Iterating backwards: The loop iterates backward through nums2 (from the last element to the first). For each element:

The stack is cleaned up by removing elements smaller than or equal to the current element (because they can't be the "next greater" element for any future elements).
If the current index i is within the original array (i <= n-1), it checks if the stack is empty:
If empty, the next greater element doesn't exist, so result[i] = -1.
Otherwise, result[i] gets the top element of the stack (the next greater element).
The current element is pushed onto the stack for future comparisons.
Return the result: The result list, which contains the next greater elements (or -1 if no greater element exists), is returned.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def nextGreaterElements(self, nums: List[int]) -> List[int]:
        
        nums2 = nums+nums
        
        s = []
        n = len(nums)
        result = [0] * n
        for i in range(len(nums2)-1, -1, -1):
            while s and s[-1] <= nums2[i]:
                s.pop()
            if i <= n-1:
                if len(s) == 0:
                    result[i] = -1
                else:
                    result[i] = s[-1]
            s.append(nums2[i])
        return result
```

<h3>3. Next Smaller Element</h3>
<a href="https://www.interviewbit.com/problems/nearest-smaller-element/discussion/p/easy-python-solution-prevsmaller/396921/993/">Problem Link</a>
<p> 
Iterate over the list A.
For each element A[i], pop elements from the stack S that are greater than or equal to A[i].
If the stack is empty, append -1 (no smaller element exists before it).
Otherwise, append the top element of the stack (the previous smaller element).
Push A[i] onto the stack for future comparisons.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    # @param A : list of integers
    # @return a list of integers
    def prevSmaller(self, A):
        S = list()
        res = []
        for i in range(len(A)):
            while (len(S) > 0 and S[-1] >= A[i]):
                S.pop()
            if (len(S) == 0):
                res.append(-1)
            else:
                res.append(S[-1])
            S.append(A[i])
        return res
```

```python
class Solution:
    def count_NGEs(self, N, arr, queries, indices):
        # Result list to store the counts for each query
        res = []
        
        # Iterate over each query index
        for idx in indices:
            count = 0
            # Compare arr[idx] with all elements to its right
            for i in range(idx + 1, N):
                if arr[i] > arr[idx]:
                    count += 1
            # Append the count for this query
            res.append(count)
        
        # Return the result list
        return res
```

<h3>5. Trapping Rainwater</h3>
<a href="https://leetcode.com/problems/trapping-rain-water/">Problem Link</a>
<p> 
n is the length of the height list, representing the number of bars.
water stores the total amount of trapped water.
stack holds indices of the bars (heights) in decreasing order.
The algorithm iterates through the list of heights with the variable right.
Inside the loop, it checks if the current bar can trap water by comparing it with the bar at the top of the stack.
If the current height is greater than the top of the stack, it pops the stack and calculates the trapped water between the current bar (right), the popped bar (mid), and the bar at the new top of the stack (left).
The water trapped between the three bars is calculated using the minimum of the left and right heights minus the height at mid, multiplied by the width between left and right.
The current bar's index is pushed onto the stack.

Check comments.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def trap(self, height: List[int]) -> int:
        n = len(height) # Number of elements in the height list
        water = 0       # Initialize the total trapped water volume
        stack = []      # Stack to store indices of height elements
        # Iterate through the heights
        for right in range(n):
            # Process each height to trap water
            while stack and height[stack[-1]] < height[right]:
                # If the current height is greater than the height at the top of the stack
                mid = stack.pop() # Get the index of the height at the top of the stack
                # If the stack becomes empty, no more water can be trapped
                if not stack:
                    break
                left = stack[-1]  # Get the index of the next height from the top of the stack
                minHeight = min(height[right] - height[mid], height[left] - height[mid]) # Calculate the minimum height of the two borders
                width = right - left - 1   # Calculate the width between the left and right borders
                water += minHeight * width # Calculate the trapped water volume and add it to the total
            stack.append(right) # Push the current index onto the stack
        return water # Return the total trapped water volume
```

<h3>6. Sum of subarray minimum</h3>
<a href="https://leetcode.com/problems/sum-of-subarray-minimums/description/">Problem Link</a>
<p> 
MOD: A large prime number to take the result modulo to prevent overflow.
stack: Used to store the indices of elements in arr to track the boundaries of the subarrays.
sumOfMinimums: Stores the running sum of the minimums of all subarrays.
The loop iterates through the array (arr), and for each element, it pops indices from the stack where the current element is smaller, computing the contribution of each popped element as the minimum of subarrays where that element is the minimum.
count: The number of subarrays where the popped element is the minimum.
Finally, the result is returned modulo MOD.

Check comments.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def sumSubarrayMins(self, arr: List[int]) -> int:
        MOD = 10**9 + 7  # Define the modulus to prevent overflow and ensure the result is within bounds
        stack = []  # Initialize an empty stack to keep track of indices
        sumOfMinimums = 0  # Variable to accumulate the sum of minimums of all subarrays

        # Loop over all elements in the array and an extra iteration to process remaining stack elements
        for i in range(len(arr) + 1):
            # While there are elements in the stack and the current element is smaller
            # than the element at the index stored at the top of the stack (or if i == len(arr) to process remaining elements)
            while stack and (i == len(arr) or arr[stack[-1]] >= arr[i]):
                mid = stack.pop()  # Pop the top element index from the stack, representing the subarray's minimum
                leftBoundary = stack[-1] if stack else -1  # The left boundary of subarrays where arr[mid] is the minimum
                rightBoundary = i  # The right boundary of subarrays where arr[mid] is the minimum

                # Calculate how many subarrays have arr[mid] as their minimum
                count = (mid - leftBoundary) * (rightBoundary - mid) % MOD

                # Add the contribution of arr[mid] to the sum, multiplied by the number of subarrays it is the minimum of
                sumOfMinimums += (count * arr[mid]) % MOD
                sumOfMinimums %= MOD  # Keep the result within the MOD value to prevent overflow

            # Push the current index onto the stack
            stack.append(i)

        return int(sumOfMinimums)  # Return the sum of minimums modulo MOD
```

<h3>7. Asteroid Collision</h3>
<a href="https://leetcode.com/problems/asteroid-collision/">Problem Link</a>
<p> 
Input: The function receives a list asteroids representing the direction and size of each asteroid.
Stack: A stack is used to track surviving asteroids as we process the list.
Iteration: For each asteroid in the list:
If it is moving right (asteroids[i] > 0), it's added directly to the stack.
If it is moving left (asteroids[i] < 0), we check if it collides with any right-moving asteroids in the stack.
Collision Handling:
If the asteroid at the top of the stack is smaller, it is popped (destroyed).
If the sizes are equal, both asteroids are destroyed.
If the top asteroid is larger, the new asteroid is destroyed.
Return: The final stack represents the surviving asteroids after all collisions.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def asteroidCollision(self, asteroids):
        n = len(asteroids)  # Get the length of the asteroids list
        stack = []  # Initialize an empty stack to store surviving asteroids
        
        # Iterate over each asteroid in the input list
        for i in range(n):
            add = True  # Flag to determine if the current asteroid should be added to the stack
            is_num_positive = asteroids[i] > 0  # Check if the current asteroid is moving to the right
            
            # Handle collisions if the current asteroid is moving left
            while stack and is_num_positive != (stack[-1] > 0) and stack[-1] > 0:
                # Check if the top asteroid in the stack is smaller
                if abs(stack[-1]) > abs(asteroids[i]):
                    add = False  # The current asteroid is destroyed
                    break  # No need to check further as this asteroid is destroyed
                
                # If both asteroids are the same size, both are destroyed
                elif abs(stack[-1]) == abs(asteroids[i]):
                    add = False  # Both asteroids are destroyed
                    stack.pop()  # Remove the asteroid from the stack
                    break  # Exit the loop as the current asteroid is also destroyed
                
                # If the asteroid in the stack is smaller, pop it off and continue checking for collisions
                else:
                    stack.pop()
            
            # If the current asteroid survived the collisions, add it to the stack
            if add:
                stack.append(asteroids[i])
        
        return stack  # Return the stack containing surviving asteroids
```

<h3>8. Sum of subarray ranges</h3>
<a href="https://leetcode.com/problems/sum-of-subarray-ranges/">Problem Link</a>
<p> 

<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def subArrayRanges(self, nums: List[int]) -> int:
        n = len(nums)
        
        # the answer will be sum{ Max(subarray) - Min(subarray) } over all possible subarray
        # which decomposes to sum{Max(subarray)} - sum{Min(subarray)} over all possible subarray
        # so totalsum = maxsum - minsum
        # we calculate minsum and maxsum in two different loops
        minsum = maxsum = 0
        
        # first calculate sum{ Min(subarray) } over all subarrays
        # sum{ Min(subarray) } = sum(f(i) * nums[i]) ; i=0..n-1
        # where f(i) is number of subarrays where nums[i] is the minimum value
        # f(i) = (i - index of the previous smaller value) * (index of the next smaller value - i) * nums[i]
        # we can claculate these indices in linear time using a monotonically increasing stack.
        stack = []
        for next_smaller in range(n + 1):
			# we pop from the stack in order to satisfy the monotonically increasing order property
			# if we reach the end of the iteration and there are elements present in the stack, we pop all of them
            while stack and (next_smaller == n or nums[stack[-1]] > nums[next_smaller]):
                i = stack.pop()
                prev_smaller = stack[-1] if stack else -1
                minsum += nums[i] * (next_smaller - i) * (i - prev_smaller)
            stack.append(next_smaller)
            
        # then calculate sum{ Max(subarray) } over all subarrays
        # sum{ Max(subarray) } = sum(f'(i) * nums[i]) ; i=0..n-1
        # where f'(i) is number of subarrays where nums[i] is the maximum value
        # f'(i) = (i - index of the previous larger value) - (index of the next larger value - i) * nums[i]
        # this time we use a monotonically decreasing stack.
        stack = []
        for next_larger in range(n + 1):
			# we pop from the stack in order to satisfy the monotonically decreasing order property
			# if we reach the end of the iteration and there are elements present in the stack, we pop all of them
            while stack and (next_larger == n or nums[stack[-1]] < nums[next_larger]):
                i = stack.pop()
                prev_larger = stack[-1] if stack else -1
                maxsum += nums[i] * (next_larger - i) * (i - prev_larger)
            stack.append(next_larger)
        
        return maxsum - minsum
```

<h3>9. Remove k Digits</h3>
<a href="https://leetcode.com/problems/remove-k-digits/description/">Problem Link</a>
<p> 
This code removes k digits from the string num such that the resulting number is the smallest possible. It uses a stack-based approach to ensure that the digits are removed in the correct order to minimize the number. Here's a breakdown:

Stack Creation: The function iterates over each digit in num, and pushes it onto the stack.
Removing Digits: While iterating, if the current digit is smaller than the top of the stack and we still have digits to remove (k > 0), the top digit is popped from the stack (this minimizes the number).
Post-Iteration Removal: If k digits haven't been removed by the end of the loop, the remaining digits are popped from the stack using another loop.
Final Processing:
After the loop, the stack is joined into a string.
Leading zeros are stripped, and if the result is empty, "0" is returned. Otherwise, the result is returned.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def removeKdigits(self, num: str, k: int) -> str:
        stack = []
        
        for char in num:
            while stack and stack[-1] > char and k: # O(K)
                k = k - 1
                stack.pop()
            
            stack.append(char)
        # Ex: num = "123456", k = 2
        # After Above For Loop, 
        # Stack = [1,2,3,4,5,6],, Nothing got Removed,, So Removal of Top K Digits Should be done by using a seperate While Loop
        while k: # O(K)
            k = k - 1
            stack.pop()   

        if ''.join(stack).lstrip('0'):
            # Ex: Stack = ['0','1','0','0']
            # ---> Joining = "0100" 
            # ---> Stripping = "100"
            return ''.join(stack).lstrip('0')
        else:
            # Ex: Stack = ['0']
            # ---> Joining = "0" 
            # ---> Stripping = "",, In this Case,, we have to Return "0"
            return '0'
```

<h3>10. Largest rectangle in a histogram</h3>
<a href="https://leetcode.com/problems/largest-rectangle-in-histogram/description/">Problem Link</a>
<p> 
Initialization: Start with a dummy index -1 in the stack to simplify width calculations.
Dummy Height: Append 0 to heights to ensure all heights are processed.
Traverse Heights:
If the current height (cur_height) is smaller than the height at the top of the stack, calculate the area for the rectangle formed with the height at the top of the stack.
Continue popping until the stack maintains a non-decreasing order of heights.
Push the current index onto the stack.
<br><br>
Time complexity: O(n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def largestRectangleArea(self, heights: List[int]) -> int:
        stack = [-1]  # Stack to store indices, starting with a dummy index (-1) for easier calculations
        heights.append(0)  # Append a dummy height (0) to process all elements and flush the stack at the end
        max_area = 0  # Variable to store the maximum rectangle area found
        
        # Iterate through each bar in the histogram (including the appended dummy height)
        for i, cur_height in enumerate(heights):
            # While the current height is less than the height at the top index of the stack
            while stack and heights[stack[-1]] > cur_height:
                height = heights[stack.pop()]  # Pop the top of the stack and get the height
                width = i - stack[-1] - 1  # Calculate the width of the rectangle using the next element in the stack
                max_area = max(max_area, width * height)  # Update the max_area if the current area is larger
            
            # Push the current index onto the stack
            stack.append(i)
        
        # Return the largest rectangle area found
        return max_area
```

<h3>11. Maximal Rectangles</h3>
<a href="https://leetcode.com/problems/maximal-rectangle/">Problem Link</a>
<p> 
Initialization:
height is an array that tracks the heights of "columns" of '1's at each column of the current row.
ans stores the maximum rectangle area found so far.

Iterate through the matrix:
For each row, update height: if row[i] == '1', increment the height; otherwise, reset it to 0.

Use a Monotonic Stack:
A stack is used to efficiently calculate the largest rectangle in the histogram formed by height.
While traversing the height array, for every bar height[i] smaller than the bar represented by the top of the stack:
Pop the stack to get the height of the rectangle.
Compute the width using the current index i and the new top of the stack.
Update ans with the maximum area.

Return the result:
After processing all rows, ans contains the largest rectangle area.
<br><br>
Time complexity: O(m*n)<br>
Space Complexity: O(n)</p>

```python
class Solution:
    def maximalRectangle(self, matrix: List[List[str]]) -> int:
        # Get the number of columns in the matrix
        n = len(matrix[0])
        
        # Initialize the height array to track "histogram heights" for each column
        # Add an extra 0 at the end for easier processing of the stack
        height = [0] * (n + 1)
        
        # Variable to store the maximum rectangle area
        ans = 0
        
        # Iterate through each row in the matrix
        for row in matrix:
            # Update the height array based on the current row
            for i in range(n):
                # Increment height if '1', otherwise reset to 0
                height[i] = height[i] + 1 if row[i] == '1' else 0
            
            # Monotonic stack to calculate the maximum rectangle area in the histogram
            stack = [-1]  # Initialize stack with a sentinel value
            
            # Traverse through the height array
            for i in range(n + 1):
                # While the current height is less than the height at the top of the stack
                while height[i] < height[stack[-1]]:
                    # Pop the stack to get the height of the rectangle
                    h = height[stack.pop()]
                    
                    # Calculate the width of the rectangle
                    w = i - 1 - stack[-1]
                    
                    # Update the maximum area
                    ans = max(ans, h * w)
                
                # Push the current index onto the stack
                stack.append(i)
        
        # Return the largest rectangle area found
        return ans
```
